```python
# This program takes the salary from the user
# Bonus amount and the final salary as given as output.
# If the salary is greater than 50000 then the bonus amount is 10 percent
# If the salary is less than 50000 then that bonus amount is 5 percent

salary = float (input ("What is your salary? "))
if salary > 50000:
  bonus = salary * 0.10
else:
  bonus = salary * 0.05

final_salary = salary + bonus

print ("Your bonus is", bonus)
print ("Your final salary is", final_salary)
```
