```python
#Question:
#Print numbers from 1 to 20
#Print only even numbers between 1 to 20

print("Numbers from 1 to 20:")
for i in range(1, 21):
    print(i)

print("\nEven numbers from 1 to 20:")
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```
